# Functions for ETL

```
import requests

def extract(url):
    f = requests.get(url)
    return (f.text)
```

```
def transform(text):
    lines = text.strip().split("\n") 
    records = []
    for i, l in enumerate(lines):
      if i == 0:
        continue
      (name, gender) = l.split(",") # l = "rkdlem196,M" -> [ 'rkdlem196', 'M' ]
      records.append([name, gender])
    return records
```
